**<a id="#">Task 4.1.2.</a>** Implement horizontal and vertical $3 \times 3$ Prewitt filters:

```python
def prewitt_h(img, size):
    result = numpy.zeros(img.shape)
    n = (size - 1) // 2
    x = numpy.array([[1, 0, -1],[1, 0, -1],[1, 0, -1]], numpy.int32)
    for p in numpy.ndindex(img.shape[1]-2*n, img.shape[0]-2*n):
        neighborhood_sum = 0
        for q in numpy.ndindex(size, size):
            pos = tuple(map(lambda x, y: x + y, p, q))
            neighborhood_sum += img[pos] * x[q]
            result[p[0]+n, p[1]+n] = neighborhood_sum/(size*2)
    return result
    
def prewitt_v(img, size):
    result = numpy.zeros(img.shape)
    n = (size - 1) // 2
    x = numpy.array([[-1, -1, -1],[0, 0, 0],[1, 1, 1]], numpy.int32)
    for p in numpy.ndindex(img.shape[1]-2*n, img.shape[0]-2*n):
        neighborhood_sum = 0
        for q in numpy.ndindex(size, size):
            pos = tuple(map(lambda x, y: x + y, p, q))
            neighborhood_sum += img[pos] * x[q]
            result[p[0]+n, p[1]+n] = neighborhood_sum/(size*2)
    return result
```

**<a id="#">Task 4.1.3.</a>** Test your above solution using the functions `prewitt_h` and `prewitt_v` for the previously loaded image and show the result:

```python
plt.imshow(prewitt_v(img.copy(), 3), 'gray')
plt.colorbar()
```
